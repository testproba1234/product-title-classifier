# 1. Uvoz biblioteka

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import matplotlib.pyplot as plt
import seaborn as sns

# Prikaz celih kolona
pd.set_option('display.max_columns', None)
```

# 2. Učitavanje podataka

```python
file_path = "https://raw.githubusercontent.com/testproba1234/product-title-classifier/refs/heads/main/data/products.csv"

import pandas as pd

df = pd.read_csv(file_path)
print("Podaci uspešno učitani!")
print(f"Broj redova: {df.shape[0]}, broj kolona: {df.shape[1]}")
df.head()
```

# 3. Pregled osnovnih informacija

```python
print("\n Informacije o kolonama:\n")
df.info()

print("\nStatistika numeričkih kolona:\n")
display(df.describe())

print("\nPrvih nekoliko redova:")
display(df.head(10))
```

# 4. Provera null vrednosti

```python
print("\nBroj praznih vrednosti po koloni:\n")
print(df.isnull().sum())

# Vizualizacija praznih vrednosti
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False, cmap="mako")
plt.title("Prikaz praznih vrednosti")
plt.show()
```

# 5. Provera duplikata

```python
dupes = df.duplicated().sum()
print(f"\nBroj duplikata u skupu podataka: {dupes}")
if dupes > 0:
    df = df.drop_duplicates()
    print("uplikati uklonjeni.")
    print(f"Novi oblik skupa: {df.shape}")
```

# 6. Distribucija kategorija

```python
if 'Category Label' in df.columns:
    plt.figure(figsize=(10,6))
    df['Category Label'].value_counts().head(20).plot(kind='bar')
    plt.title("Top 20 kategorija po učestalosti")
    plt.xlabel("Kategorija")
    plt.ylabel("Broj proizvoda")
    plt.show()
else:
    print("Kolona 'Category Label' nije pronađena — proveri naziv kolone!")
```

# 7. Analiza dužine naslova proizvoda

```python
df['title_length'] = df['Product Title'].astype(str).apply(len)
df['word_count'] = df['Product Title'].astype(str).apply(lambda x: len(x.split()))

plt.figure(figsize=(10,5))
sns.histplot(df['title_length'], bins=50, kde=True)
plt.title("Distribucija dužine naslova proizvoda (broj karaktera)")
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(df['word_count'], bins=30, kde=True)
plt.title("Distribucija broja reči u naslovu proizvoda")
plt.show()

print("\nProsečna dužina naslova:", df['title_length'].mean())
print("Prosečan broj reči u naslovu:", df['word_count'].mean())
```

# 8. Provera brendova i specijalnih reči

```python
df['has_brand'] = df['Product Title'].apply(lambda x: bool(re.search(r'\b(Apple|Samsung|Sony|LG|HP|Asus|Lenovo|Dell|Canon|Nikon)\b', str(x), re.I)))
df['contains_number'] = df['Product Title'].apply(lambda x: bool(re.search(r'\d', str(x))))
df['contains_upper'] = df['Product Title'].apply(lambda x: any(word.isupper() for word in str(x).split()))

print("\nPrimeri sa detektovanim brendom:")
display(df[df['has_brand']].head(10))
```

# 9. Korelacije i dodatne inspekcije

```python
num_cols = df.select_dtypes(include=[np.number])
if not num_cols.empty:
    plt.figure(figsize=(8,5))
    sns.heatmap(num_cols.corr(), annot=True, cmap="coolwarm")
    plt.title("Korelacija numeričkih kolona")
    plt.show()
else:
    print("Nema numeričkih kolona za korelaciju.")
```
